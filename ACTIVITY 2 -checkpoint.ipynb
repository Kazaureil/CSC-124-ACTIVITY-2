```python
import time
import numpy as np
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt

#Recursion
def recursion_fibo(n):
    
    
   if n <= 1:
       return n
   else:
       return(recursion_fibo(n-1) +recursion_fibo(n-2))

numfterms = 10
tStart1 = time.time()
# check if the number of terms is valid
if numfterms <= 0:
   print("Plese enter a positive integer")
else:
   print("Fibonacci sequence:")
   for i in range(numfterms):
    
       print(recursion_fibo(i))
        
tEnd1 = time.time()
runTime1 = tEnd1 - tStart1
print(runTime1)
  

    
#Iteration
def iterate():


	

	a = 0
	b = 1
	terms = 10
	print(a,b,end=" ")
	while(terms-2):
		c = a + b
		a,b = b,c
		print(c,end=" ")
		terms = terms - 1
tStart2 = time.time()
iterate()
tEnd2 = time.time()
runTime2 = tEnd2 - tStart2
print(runTime2)





#REGULAR
numfterms = int(input("How many terms? "))
# first two terms
n1, n2 = 0, 1
count = 0
tStart3 = time.time()
# check if the number of terms is valid
if numfterms <= 0:
   print("Please enter a positive integer")
elif numfterms == 1:
   print("Fibonacci sequence upto",numfterms,":")
   print(n1)
else:
   print("Fibonacci sequence:")
   while count < numfterms:
       print(n1)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
tEnd3 = time.time()
runTime3 = tEnd3 - tStart3
print(runTime3)



objects = ('RECURSION', 'ITERATION', 'ONMYOWN')
y_pos = np.arange(len(objects))
performance = [runTime1,runTime2,runTime3]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('SPEED')
plt.title('FIBONACCI')

plt.show()
```
